```python
visa_df=pd.read_csv(r"C:\Users\User\Desktop\Naresh IT\datafiles\Visadataset.csv")
```

```python
visa_df['case_status'].unique()
```

```python
cols=visa_df.select_dtypes(include='object').columns


for i in cols:
    labels=list(visa_df[i].unique())
    shubham=[i for i in range(len(labels))]
    d=dict(zip(labels,shubham))
    visa_df[i]=visa_df[i].map(d)

visa_df
```

```python
cols=visa_df.select_dtypes(include='object').columns

for i in cols:
    labels=list(visa_df[i].unique())
    values=[i for i in range(len(labels))]
    d=dict(zip(labels,values))
    visa_df[i]=visa_df[i].map(d)

visa_df
```
