**Create data frame using dictionary**

```python
name=['avinash','aakash','aadhya']
age=[29,30,23]
pd.DataFrame(zip(name,age),columns=['Names','Age'])
```

```python
dict1={'Names':['avinash','aakash','aadhya'],
      'Ages':[29,30,23]}
dict1
```

```python
pd.DataFrame(dict1)

#keys becomes columns
#values becomes rows 
#no need to provide colums name seperately
```

```python
dict2={'Names':'aadhya',
      'Ages':23,
      'City':'hyd'}
dict2
```

```python
pd.DataFrame(dict2,index=[1,2,3])
```
