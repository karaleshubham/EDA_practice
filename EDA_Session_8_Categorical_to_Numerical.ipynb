## Label Encoding 
**Label Encoding**g is a technique that is used to convert categorical columns into numerical one

 so that they can be fitted by machine learning models which only take numerical data

 It is an important pre-processing step in a machine-learning project.

- Label encoder is a package under **sickit learn**

- Sickit-learn is is heart of Machine learning

- In side sickit learn we have method called pre-processing

- underpreprocessing we have label encoder**

- sickit -learn we will write as sklearn 

- any sklearn package we have 3 steps 

    - Step-1: Read the package 

    - Step-2:save the package

**One hot encoder**

- One hot means if one is ON another one is OFF

- ON represents with 1, and OFF represents with 0

- for exapmle case staus has two unique lables

    - Denied 

    - Certified

- If you apply one hot encoder on case status column it will create two extra columns 

    - case_status_Denied

    - case_status_Certified



```python
|   case_status    |   case_status_denied   |    case_status_Certified    |
==========================================================================
|  Denied           |           1            |               0            |
--------------------------------------------------------------------------
|  Certified        |           0            |               1             |
```

**Ad

-  In machine learning we have one important property is **Independent of variables**

- Machine Learning models expects variables are **Independent each other**

- One hot encoder crearting new columns, the new columns are **Independent each other**

- Independent variables means both variables have **90 degrees phase shift**

- 90 degreesw phase shift also called as **perpendicular each other**

- Perpendicular means **orthogonal each other**

**Disadvantages**

- One hot encoder creates new columns equal to number of unique labels

- If number of unique labels increase , new columns also increase

- The more columns leads data is more 

- Then data is more the complexity will increase

- Complecity increases more computation power required 

- Also time also increse

- The entire process named as **curse of dimenionlity**
