***Task***

- Create categorical list

- Create numerical column list

**idea**

- convert series type into dictionary 

- iterate through loop 

- apply the condition get the result in the list 


```python
dict(visa_df.dtypes)
```

```python
dict2=dict(visa_df.dtypes)

for i in dict2:
    
    if dict2[i]==object:
        print("categorical")
    elif dict2[i]==int or dict2[i]==float:
        print("numerical")
    
    else:
        print("nothing")
```

```python
for key,value in dict2.items():
    if value=="object":
        print(f"{key}:{value}")
    
    else:
        print(f"{key}:{value}")
```

$Take$

- take is used to extract the perticular index of data

- take has one parameter called axis

- axis=1 means columns

- axis=0 means rows

- by default axis=0 is there

- in below example we given 100,200,300 and we did not provide axis values

- it understood we are talking about rows

-  in take we cant get the rows n colums at a time #drawback

```python
visa_df.take([100,200,300])
```

```python
visa_df.take([11],axis=1)# only 12 rows are there so we cant use more than that or else it will give error
```

```python
visa_df.take([i for i in range (200)]) #instead of that we can use slicing df[1:200]
```
